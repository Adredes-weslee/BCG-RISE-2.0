### Question 1: Check if a number is Palindrome* [10%]

    *A palindrome number is a number that remains the same when its digits are reversed. In other words, it reads the same backward as forward. Palindrome numbers are symmetric and exhibit the same sequence of digits when read from left to right and from right to left.

<b>a. Write a code snippet for user to input a number. [2%]</b>

<b>b. Run a check to see if the input is a number or a string, if user input a string, display out a message showing, "Incorrect format, please input a number." [4%]</b>

<b>c. Once the user inputs a number, run a check to see if the number is palindrome or not. If the number is palindrome display "The given number is Palindrome". If it's not a palindrome, then display, "The given number is not a Palindrome. [4%]</b>

##### Examples:

input: "bib"<br>
output: Incorrect format, please input a number.

input: "415"<br>
output: The given number is not a Palindrome.

input: "212"<br>
output: The given number is Palindrome.




```python
# Prompt user to input a number
user_input = input("Please enter a number: ")

# Check if the input is a number or a string
try:
    # Try to convert the input to a float to handle numbers (both integers and decimals)
    number = float(user_input)
    
    # Check if the number is a decimal by comparing it to its integer part
    if number != int(number):
        print("Incorrect format, please input an integer.")
    else:
        # If the number is negative, take the absolute value and print a warning
        if number < 0:
            print("Warning: Negative number detected. Taking the absolute value.")
            number = abs(number)
        
        # Convert the number to an integer (since we know it's not a decimal at this point)
        number = int(number)
        
        # Check if the number is a palindrome
        # A number is a palindrome if it reads the same forward and backward
        reversed_number = int(str(number)[::-1])  # Reverse the number by converting it to a string, reversing, and converting back to int
        if number == reversed_number:
            print("The given number is Palindrome.")
        else:
            print("The given number is not a Palindrome.")
except ValueError:
    # If conversion fails, it means the input is not a number
    print("Incorrect format, please input a number.")
```

### Question 2:  Remove STOP words* [10%]

    *In NLP, stop words are commonly used words like "a", "is", and "the". 
    They are typically filtered out during processing.

<b>a. Implement a function that takes a string text and an integer k. [2%]</b>

<b>b. Checks the occurrence of each word in the sentence. [3%]</b>

<b>c. Returns the list of words that occur in the text at least k times. [3%]</b>

<b>d. The words must be returned in the order of their first occurrence in the text. [2%]</b>

##### Examples:

input:

    text = "a mouse is smaller than a dog but a dog is stronger"
    k=2

output: 

    ['a', 'is', 'dog']

##### Explanation: 
- The list of stop words that occur at least k = 2Â times is ["a", "is", "dog"]. 
- "a" occurs 3 times, "is" and "dog" both occur 2 times. No other word occurs at least 2 times. 
- The answer is in order of the first appearance in text.




```python
def words_at_least_k_times(text, k):
    # Split the text into words
    words = text.split()  # Split the text string into a list of words based on spaces
    
    # Dictionary to store the count of each word
    word_count = {}  # Initialize an empty dictionary to count occurrences of each word
    
    # List to keep the order of the first occurrence
    first_occurrence = []  # Initialize an empty list to maintain the order of first occurrence of each word
    
    # Count the occurrence of each word
    for word in words:
        if word not in word_count:
            word_count[word] = 1  # If the word is not in the dictionary, add it with a count of 1
            first_occurrence.append(word)  # Add the word to the list of first occurrences
        else:
            word_count[word] += 1  # If the word is already in the dictionary, increment its count
    
    # List to store words that occur at least k times
    result = [word for word in first_occurrence if word_count[word] >= k]  # Create a list of words that appear at least k times
    
    return result  # Return the final list of words
```

```python
# Example usage
text = "a mouse is smaller than a dog but a dog is stronger"
k = 2
output = words_at_least_k_times(text, k)
print(output)  # Output: ['a', 'is', 'dog']
```

### Question 3: Suffix stripping Stemmer* [10%]

    *Stemming is the process of extracting the base word from a word. For instance, the base for "worked" is "work".
    
<b>a) Implement a function that takes a sentence without punctuation. [1%]</b>

<b>b) If the word ends in 'ed', 'ly' or 'ing', remove the suffix. [4%]</b>

<b>c) If the resulting word is longer than 7 letters, keep the first 7 letters. [3%]</b>

<b>d) Return the stemmed counterpart of individual words. [2%]</b>

##### Example:

input:

    text = "an extremely ferocious dog is barking"

output: 

    an extreme ferocio dog is bark

##### Explanation: 
- 'an' does not end in one of the suffixes and is less than 8 letters
- 'extremely' is 'extreme' after removing the suffix and 'extreme' is less than 8 letters
- 'ferocious' is 9 letters long, so reduce it to 7 letters
- ' ferocio','dog' and 'is' are unchanged
- 'barking' is 'bark' after removing the suffix and is less than 7 letters

```python
def stem_sentence(text):
    # Split the text into words
    words = text.split()  # Split the input text into a list of words based on spaces
    
    # Function to stem a single word
    def stem_word(word):
        # Remove the suffix 'ed' if present
        if word.endswith('ed'):
            word = word[:-2]
        # Remove the suffix 'ly' if present
        elif word.endswith('ly'):
            word = word[:-2]
        # Remove the suffix 'ing' if present
        elif word.endswith('ing'):
            word = word[:-3]
        
        # If the resulting word is longer than 7 letters, truncate it to the first 7 letters
        if len(word) > 7:
            word = word[:7]
        
        return word
    
    # Apply stemming to each word in the sentence
    stemmed_words = [stem_word(word) for word in words]  # Use a list comprehension to stem each word
    
    return ' '.join(stemmed_words)  # Join the list of stemmed words back into a single string
```

```python
# Example usage
text = "an extremely ferocious dog is barking"
output = stem_sentence(text)
print(output)  # Output: "an extreme ferocio dog is bark"
```

### Question 4: Scientific Calculator [10%]

<b>a) Implement a function that to create a simple scientific calculator. [2%]</b>

<b>b) The operations that can be performed are: [7%]

    - Addition
    - Subtraction
    - Multiplication
    - Division
    - Exponential
    - Square root
    - Logarithm
    - Sin
    - Cosine
    - Tan
</b>
<b>c) Return the results after computing the selected operation. [1%]</b>

```python
import math

def scientific_calculator(operation, *args):
    # Check if the operation is 'add'
    if operation == 'add':
        # Return the sum of all arguments
        return sum(args)
    # Check if the operation is 'subtract'
    elif operation == 'subtract':
        # Subtract all subsequent arguments from the first argument
        return args[0] - sum(args[1:])
    # Check if the operation is 'multiply'
    elif operation == 'multiply':
        # Initialize the result to 1
        result = 1
        # Multiply each argument with the result
        for num in args:
            result *= num
        return result
    # Check if the operation is 'divide'
    elif operation == 'divide':
        try:
            # Initialize the result with the first argument
            result = args[0]
            # Divide the result by each subsequent argument
            for num in args[1:]:
                result /= num
            return result
        except ZeroDivisionError:
            # Handle division by zero
            return "Error: Division by zero"
    # Check if the operation is 'exponential'
    elif operation == 'exponential':
        # Return the first argument raised to the power of the second argument
        return args[0] ** args[1]
    # Check if the operation is 'sqrt'
    elif operation == 'sqrt':
        # Return the square root of the first argument
        return math.sqrt(args[0])
    # Check if the operation is 'log'
    elif operation == 'log':
        # Return the natural logarithm of the first argument
        return math.log(args[0])
    # Check if the operation is 'sin'
    elif operation == 'sin':
        # Return the sine of the first argument (converted to radians)
        return math.sin(math.radians(args[0]))
    # Check if the operation is 'cos'
    elif operation == 'cos':
        # Return the cosine of the first argument (converted to radians)
        return math.cos(math.radians(args[0]))
    # Check if the operation is 'tan'
    elif operation == 'tan':
        # Return the tangent of the first argument (converted to radians)
        return math.tan(math.radians(args[0]))
    # If the operation is not recognized
    else:
        return "Invalid operation"
```

```python
# Example usage
print(scientific_calculator('add', 1, 2, 3))           # Output: 6
print(scientific_calculator('subtract', 10, 2, 3))     # Output: 5
print(scientific_calculator('multiply', 2, 3, 4))      # Output: 24
print(scientific_calculator('divide', 20, 4, 2))       # Output: 2.5
print(scientific_calculator('exponential', 2, 3))      # Output: 8
print(scientific_calculator('sqrt', 16))               # Output: 4.0
print(scientific_calculator('log', 100))               # Output: 4.605170185988092
print(scientific_calculator('sin', 30))                # Output: 0.49999999999999994
print(scientific_calculator('cos', 60))                # Output: 0.5000000000000001
print(scientific_calculator('tan', 45))                # Output: 0.9999999999999999
print(scientific_calculator('test', 45))               # Output: Invalid operation
```

```python
# Select relevant columns
df_products = df_temp[['orderid', 'item_name']]

# Pivot the dataframe to have orderid as index and item_name as columns
df_pivot = df_products.pivot_table(index='orderid', columns='item_name', aggfunc=lambda x: 1, fill_value=0)

# Convert all non-zero values to 1 using applymap
df_encoded = df_pivot.apply(lambda x: x.map(lambda x: 1 if x > 0 else 0))

# Print df_encoded
df_encoded
```

### Building a Library Management System [10%]

You are tasked with building a library management system using Python classes and objects. The system should allow for the management of books, library members, and borrowing transactions. Here are the requirements:

1. <b>Create a class called Book with the following attributes and methods: [2%] </b>

    `Attributes:` title, author, isbn, copies_available, total_copies

    `Method:` display_info() that displays information about the book.
    

2. <b>Create a class called Member with the following attributes and methods: [3%] </b>

    `Attributes:` name, member_id, books_borrowed (a list of books borrowed)

    `Method:` borrow_book(book) that allows a member to borrow a book.

    `Method:` return_book(book) that allows a member to return a book.
    

3. <b>Create a class called Library with the following attributes and methods: [5%] </b>

    `Attributes:` name, books (a list of books), members (a list of members)

    `Method:` add_book(book) to add a book to the library.

    `Method:` add_member(member) to add a member to the library.

    `Method:` get_available_books() to get a list of books available for borrowing.

    `Method:` get_borrowed_books() to get a list of borrowed books.

    `Method:` get_member_books(member) to get a list of books borrowed by a member.

Your goal is to create the above class hierarchy and implement the methods to demonstrate interactions between books, members, and the library. 

Test your classes by creating instances, adding books and members, and simulating borrowing and returning transactions.

##### Example outputs:
    John Smith has borrowed 'The Great Gatsby'.
    Alice Johnson has borrowed 'The Great Gatsby'.
    John Smith has borrowed 'To Kill a Mockingbird'.
    John Smith has returned 'The Great Gatsby'.
    Available Books: ['The Great Gatsby', 'To Kill a Mockingbird']
    Borrowed Books: ['The Great Gatsby', 'To Kill a Mockingbird']
    Books borrowed by John Smith: ['To Kill a Mockingbird']

```python
class Book:
    def __init__(self, title, author, isbn, total_copies):
        """
        Initialize a new Book instance.
        
        Parameters:
        title (str): The title of the book.
        author (str): The author of the book.
        isbn (str): The ISBN number of the book.
        total_copies (int): The total number of copies available for this book.
        """
        self.title = title
        self.author = author
        self.isbn = isbn
        self.total_copies = total_copies
        self.copies_available = total_copies  # Initially, all copies are available

    def display_info(self):
        """
        Display information about the book.
        """
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"ISBN: {self.isbn}")
        print(f"Copies Available: {self.copies_available}/{self.total_copies}")

class Member:
    def __init__(self, name, member_id):
        """
        Initialize a new Member instance.
        
        Parameters:
        name (str): The name of the member.
        member_id (str): The ID of the member.
        """
        self.name = name
        self.member_id = member_id
        self.books_borrowed = []  # List to keep track of borrowed books

    def borrow_book(self, book):
        """
        Borrow a book from the library.
        
        Parameters:
        book (Book): The book to be borrowed.
        """
        if book.copies_available > 0:
            self.books_borrowed.append(book)  # Add book to the member's borrowed list
            book.copies_available -= 1  # Decrease the number of available copies
            print(f"{self.name} has borrowed '{book.title}'.")
        else:
            print(f"Sorry, '{book.title}' is not available for borrowing.")

    def return_book(self, book):
        """
        Return a borrowed book to the library.
        
        Parameters:
        book (Book): The book to be returned.
        """
        if book in self.books_borrowed:
            self.books_borrowed.remove(book)  # Remove book from the member's borrowed list
            book.copies_available += 1  # Increase the number of available copies
            print(f"{self.name} has returned '{book.title}'.")
        else:
            print(f"{self.name} has not borrowed '{book.title}'.")

class Library:
    def __init__(self, name):
        """
        Initialize a new Library instance.
        
        Parameters:
        name (str): The name of the library.
        """
        self.name = name
        self.books = []  # List to store all books in the library
        self.members = []  # List to store all members of the library

    def add_book(self, book):
        """
        Add a new book to the library.
        
        Parameters:
        book (Book): The book to be added to the library.
        """
        self.books.append(book)  # Add the book to the library's book list
        print(f"'{book.title}' has been added to the library.")

    def add_member(self, member):
        """
        Add a new member to the library.
        
        Parameters:
        member (Member): The member to be added to the library.
        """
        self.members.append(member)  # Add the member to the library's member list
        print(f"Member '{member.name}' with ID '{member.member_id}' has been added to the library.")

    def get_available_books(self):
        """
        Get a list of books that are currently available for borrowing.
        
        Returns:
        list: A list of books that have available copies.
        """
        available_books = [book for book in self.books if book.copies_available > 0]
        return available_books

    def get_borrowed_books(self):
        """
        Get a list of books that are currently borrowed.
        
        Returns:
        list: A list of books that are currently borrowed.
        """
        borrowed_books = [book for book in self.books if book.copies_available < book.total_copies]
        return borrowed_books

    def get_member_books(self, member):
        """
        Get a list of books that a specific member has borrowed.
        
        Parameters:
        member (Member): The member whose borrowed books are to be retrieved.
        
        Returns:
        list: A list of books borrowed by the specified member.
        """
        member_books = [book for book in member.books_borrowed]
        return member_books
```

```python
# Create a library instance
library = Library("My Library")

# Create book instances
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "978-1-234567-89-0", 2)
book2 = Book("To Kill a Mockingbird", "Harper Lee", "978-3-456789-12-0", 1)
book3 = Book("1984", "George Orwell", "978-0-123456-78-9", 3)
book4 = Book("Pride and Prejudice", "Jane Austen", "978-1-987654-32-1", 2)

# Create member instances
member1 = Member("John Smith", "M001")
member2 = Member("Alice Johnson", "M002")
member3 = Member("Robert Brown", "M003")

# Add books to the library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)

# Add members to the library
library.add_member(member1)
library.add_member(member2)
library.add_member(member3)

# Borrow books
member1.borrow_book(book1)
member2.borrow_book(book1)
member1.borrow_book(book2)
member3.borrow_book(book3)
member3.borrow_book(book4)

# Return a book
member1.return_book(book1)

# Get available books
available_books = library.get_available_books()
print("Available Books:", [book.title for book in available_books])

# Get borrowed books
borrowed_books = library.get_borrowed_books()
print("Borrowed Books:", [book.title for book in borrowed_books])

# Get books borrowed by members
member_books = library.get_member_books(member1)
print(f"Books borrowed by {member1.name}:", [book.title for book in member_books])

member_books = library.get_member_books(member2)
print(f"Books borrowed by {member2.name}:", [book.title for book in member_books])

member_books = library.get_member_books(member3)
print(f"Books borrowed by {member3.name}:", [book.title for book in member_books])
```
